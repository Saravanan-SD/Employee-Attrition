```python
Emp_dict = attrdata["Emp. Group"].value_counts()
Emp_dict['other group'] = 1
print(Emp_dict)
```

```python
job_dict = attrdata["Job Role Match"].value_counts()
job_dict
```

```python
tenure_dict = attrdata["Tenure Grp."].value_counts()
print(tenure_dict)
```

```python
# Build new dictionary to be used to categorize data column after
location_dict = attrdata["Location"].value_counts()
print(location_dict)

location_dict_new = {
    'Chennai':       7,
    'Noida':         6,
    'Bangalore':     5,
    'Hyderabad':     4,
    'Pune':          3,
    'Madurai':       2,
    'Lucknow':       1,
    'other_place':         0,
}

print(location_dict_new)
```

```python
def location(x):
    if str(x) in location_dict_new.keys():
        return location_dict_new[str(x)]
    else:
        return location_dict_new['other_place']
    
attrdata["New Location"]= attrdata["Location"].apply(location)
attrdata.head()
```

```python
fn= pd.get_dummies(attrdata["Function"]).astype(int)
Hiring_source= pd.get_dummies(attrdata["Hiring Source"]).astype(int)
 
```

```python
def Mar(x):
    if str(x) in Marital_dict.keys() and Marital_dict[str(x)]>100:
        return str(x)
    else:
        return 'other status'
    
attrdata["New Marital"]= attrdata["Marital Status"].apply(Mar)
attrdata.head()
```

```python
Mr=pd.get_dummies(attrdata['New Marital']).astype(int)
```

```python
def Promoted(x):
    if x=='Promoted':
        return int(1)
    else:
        return int(0)
    
attrdata["New Promotion"]= attrdata["Promoted/Non Promoted"].apply(Promoted)
attrdata.head()
```

```python
#Employee Group

Emp_dict_new ={
    'B1': 4,
    'B2': 3,
    'B3': 2,
    'other group':1,
}

def emp(x):
    if str(x) in Emp_dict_new.keys():
        return Emp_dict_new[str(x)]
    else:
        return Emp_dict_new['other group']

attrdata["New EMP"]= attrdata["Emp. Group"].apply(emp)
attrdata.head()
```

```python
def job(x):
    if x== 'Yes':
        return int(1)
    else:
        return int(0)
    
attrdata["New Job role match"]= attrdata["Job Role Match"].apply(job)
attrdata.head()
```

```python
gend = pd.get_dummies(attrdata["Gender "]).astype(int)

tenure=pd.get_dummies(attrdata["Tenure Grp."]).astype(int)
```

```python
df=pd.concat([attrdata,fn,Hiring_source,Mr,gend,tenure],axis=1)
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop(['name', 'phone number', 'Location', 'Emp. Group', 'Function', 'Gender ',
       'Tenure', 'Tenure Grp.','Marital Status','Hiring Source', 'Promoted/Non Promoted',
       'Job Role Match','New Marital'],axis=1,inplace=True)
```

```python
df.info()
```

```python
target=pd.get_dummies(df["Stay/Left"]).astype(int)
```

```python
def target(x):
    if x== 'Stay':
        return int(1)
    else:
        return int(0)
    
df["Target"]= df['Stay/Left'].apply(target)
df.head()
```

```python
X=  df.drop(['Stay/Left','Target'],axis=1)
y= df['Target']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
lr= LogisticRegression(C=0.1,random_state=42,solver='liblinear')
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
gnb=GaussianNB()
```

```python
for a,b in zip([lr,dt,rf,gnb],["Logistic Regression","Decision Tree","Random Forest","Naive Bayes"]):
    a.fit(X_train,y_train)
    test_pred=a.predict(X_test)
    train_pred=a.predict(X_train)
    train_score=accuracy_score(y_train,train_pred)
    test_score=accuracy_score(y_test,test_pred)
    print(f'{b} Training accuracy is: {train_score}')
    print(f'{b} Testing accuracy is: {test_score}')
```

```python
model=lr.fit(X_train,y_train)

import pickle
filename= 'Attrition_model.pickle'
pickle.dump(model,open(filename,'wb'))
```
